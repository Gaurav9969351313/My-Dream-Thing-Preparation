```python
import requests
import pandas as pd
import numpy as np

date = "04102022"
daily_security_wise_dilivery_report_url = 'https://archives.nseindia.com/archives/equities/mto/MTO_'+date+'.DAT'
report_file = requests.get(daily_security_wise_dilivery_report_url)
path = './' + date + '.DAT'
open(path, 'wb').write(report_file.content)

topex = np.fromfile(path)
df_topex = pd.DataFrame(data=topex)
df_topex.head(10)
```
